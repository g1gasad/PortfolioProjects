```python
#importing libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8)  #adjusts the configuration of the plots we'll create

df = pd.read_csv('E:\Raw Data\movie industry\movies.csv')
```

```python
df.head()
df.shape
```

```python
# looking if there's any missing data
for col in df.columns:
    missing = (df[col].isnull().sum())
    print(f'{col} : {missing}')
```

```python
#dropping na values or null values
df = df.dropna()
print(df.shape)
```

```python
# checking if there's any missing data left
for col in df.columns:
    missing = (df[col].isnull().sum())
    print(f'{col} : {missing}')
```

```python
df.dtypes
```

```python
# converting float dtypes to int
df['votes'] = df["votes"].astype('int64')
df['budget'] = df["budget"].astype('int64')
df['gross'] = df["gross"].astype('int64')
df['runtime'] = df["runtime"].astype('int64')
df['year'] = df['year'].astype('int64')
```

```python
df
```

```python
# I have left one code cell empty for year and release column
```

```python
# sorting by gross revenue
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# to display all of the data
pd.set_option('display.max_rows', None)
```

```python
pd.reset_option('display.max_rows')
```

```python
# drop any duplicates
df.drop_duplicates()
# this gives us the distinct company values
# df.drop_duplicates().sort_values(ascending=False)
```

```python
# scatter plot with budget v gross
plt.scatter(x=df['budget'], y=df['gross'])
plt.title("Budget v Gross")
plt.xlabel('Budget')
plt.ylabel("Gross Revenue")

plt.show()
```

```python
# we're going to plot budget v gross using sns
sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'green'}, line_kws={'color':'cyan'})
```

```python
df_numerized = df
```

```python
for col_name in df_numerized.columns:
    if df_numerized[col_name].dtype == 'object':
        df_numerized[col_name] = df_numerized[col_name].astype("category")
        df_numerized[col_name] = df_numerized[col_name].cat.codes
# converting object to cat dtypes
# cat.codes gives certain codes to data
df_numerized
```

```python
# heatmap for all the numerize data
corr_matrix = df_numerized.corr(method='pearson')

sns.heatmap(corr_matrix, annot=True)
plt.title("Correlation Matrix for Numeric Features")
plt.xlabel('Movie Features')
plt.ylabel("Movie Features")
plt.show()
```

```python
df_numerized.corr()
```

```python
corr_matrix = df_numerized.corr()

corr_pairs = corr_matrix.unstack()
pd.set_option('display.max_rows',None)
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[sorted_pairs > 0.5]
high_corr
```
